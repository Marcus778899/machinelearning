```python
import numpy as np
from sklearn import datasets,linear_model
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.model_selection import train_test_split
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
diabetes = datasets.load_diabetes()
diabetes_x = diabetes.data
diabetes_y = diabetes.target
print(diabetes_y.shape)
print(diabetes_x.shape)
```

```python
x_train,x_test,y_train,y_test = train_test_split(diabetes_x,diabetes_y,test_size=0.1,random_state=1)
model = linear_model.LinearRegression(n_jobs=1)
model.fit(x_train,y_train)
pred = model.predict(x_test)
print(model.coef_)
print(mean_squared_error(y_test,pred))
print(r2_score(y_test,pred))
```

```python
dia_x = diabetes.data[:,None,3]
dia_y = diabetes.target
print(dia_x.shape)
dia_x_train,dia_x_test,dia_y_train,dia_y_test = train_test_split(dia_x,dia_y,test_size=0.1,random_state=1)
model = linear_model.LinearRegression(n_jobs=3)
model.fit(dia_x_train,dia_y_train)
predict = model.predict(dia_x_test)
print(model.coef_)
print(mean_squared_error(dia_y_test,predict))
print(r2_score(dia_y_test,predict))
plt.scatter(dia_x_test,dia_y_test,color = 'red')
plt.plot(dia_x_test,predict)
```
